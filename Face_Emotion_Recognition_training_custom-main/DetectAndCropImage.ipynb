```python
import cv2
import os

# Load the pre-trained face detection model
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

# Input and output directories
# input_dir = '/content/drive/MyDrive/private_test'
input_dir = '/content/drive/MyDrive/public_test'
# output_dir = '/content/drive/MyDrive/test'

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Function to process each image
def process_image(filename):
    try:
        # Load the image
        image_path = os.path.join(input_dir, filename)
        image = cv2.imread(image_path)

        if image is None:
            raise Exception(f"Error loading image: {filename}")

        # Convert the image to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Perform face detection
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # Draw rectangles around the detected faces
        for (x, y, w, h) in faces:
            cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # Save the image with bounding boxes to the output directory
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, image)

        print(f"Processed: {filename}")

    except Exception as e:
        print(f"Error processing {filename}: {str(e)}")

# Process each image in the input directory
try:
    for filename in os.listdir(input_dir):
        # Check if the file is an image
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            process_image(filename)
except Exception as e:
    print(f"Error listing files in the input directory: {str(e)}")
```
